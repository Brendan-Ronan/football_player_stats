```python
import pandas as pd
import numpy as np
import re

df = pd.read_csv("cleansed_players.csv")
df.head()
```

```python
df.rename(columns = {
    "player_name": "Player Name",
    "position": "Position",
    "team": "Team",
    "college": "College"
}, inplace=True)

df.drop_duplicates()
```

```python
regex = r'(?P<Last_Name>.*),\s(?P<First_Name>.*)'

df[['Last_Name', 'First_Name']] = df['Player Name'].str.extract(regex)

df = df.drop('Player Name', axis=1)

df.head(10)
```

```python
df.to_csv('nfl_colleges_analysis.csv', index=False)
```
